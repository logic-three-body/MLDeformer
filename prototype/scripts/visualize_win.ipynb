# MLDeformer Prototype Inference Visualization\n\n读取 `infer_report.json` 并展示误差/时延对比。

```python
from pathlib import Path\nimport json\nimport pandas as pd\nimport matplotlib.pyplot as plt\n\nDATA_ROOT = Path('D:/MLDPrototypeData')\nindex_path = DATA_ROOT / 'artifacts' / 'infer' / 'infer_reports_index.json'\nindex = json.loads(index_path.read_text(encoding='utf-8'))\nrows = []\nfor item in index.get('reports', []):\n    report = json.loads(Path(item['infer_report']).read_text(encoding='utf-8'))\n    rows.append(report)\ndf = pd.DataFrame(rows)\ndf
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))\ndf.plot(x='method', y='mean_error', kind='bar', ax=axes[0], legend=False, color='#2C7FB8')\naxes[0].set_title('Mean Error by Method')\naxes[0].set_ylabel('Mean Error')\n\ndf.plot(x='method', y='mean_latency_ms', kind='bar', ax=axes[1], legend=False, color='#F03B20')\naxes[1].set_title('Latency (ms) by Method')\naxes[1].set_ylabel('ms')\n\nplt.tight_layout()\nplt.show()
```
